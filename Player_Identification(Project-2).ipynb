```python
import cv2
from ultralytics import YOLO
from sort.sort import Sort
import numpy as np

# Load YOLO model
model = YOLO(r"C:\Users\Arunava Chakraborty\Downloads\best.pt")

# Initialize SORT tracker
tracker = Sort()

# Load video
cap = cv2.VideoCapture(r"C:\Users\Arunava Chakraborty\Downloads\15sec_input_720p.mp4")

# Get video properties safely
fps = cap.get(cv2.CAP_PROP_FPS)
if fps == 0 or fps is None:
    fps = 30  # fallback

width = int(cap.get(cv2.CAP_PROP_FRAME_WIDTH))
height = int(cap.get(cv2.CAP_PROP_FRAME_HEIGHT))

print(f"Video properties: {fps} FPS, {width}x{height}")

# Create VideoWriter object
fourcc = cv2.VideoWriter_fourcc(*'mp4v')
out = cv2.VideoWriter('output_video.mp4', fourcc, fps, (width, height))

frame_count = 0

while True:
    ret, frame = cap.read()
    if not ret:
        break

    frame_count += 1

    # Resize frame if needed to match output dimensions
    frame = cv2.resize(frame, (width, height))

    # Run YOLO detection
    results = model(frame)

    detections = []
    for r in results[0].boxes:
        x1, y1, x2, y2 = r.xyxy[0].cpu().numpy()
        score = r.conf[0].cpu().numpy()
        if score > 0.5:  # confidence threshold
            detections.append([x1, y1, x2, y2, score])

    detections = np.array(detections)

    print(f"Frame {frame_count}: {len(detections)} detections")

    # Update tracker
    if detections.shape[0] == 0:
        tracked_objects = tracker.update(np.empty((0, 5)))
    else:
        tracked_objects = tracker.update(detections)

    # Draw tracked objects
    for obj in tracked_objects:
        x1, y1, x2, y2, track_id = obj
        cv2.rectangle(frame, (int(x1), int(y1)), (int(x2), int(y2)), (0, 255, 0), 2)
        cv2.putText(frame, f'ID: {int(track_id)}', (int(x1), int(y1) - 10),
                    cv2.FONT_HERSHEY_SIMPLEX, 0.6, (0, 255, 0), 2)

    # Write processed frame to output
    out.write(frame)

# Release resources
cap.release()
out.release()
cv2.destroyAllWindows()

print(" Processing complete. Output saved as output_video.mp4")
```
